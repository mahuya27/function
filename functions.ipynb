##1. What is the difference between a function and a method in Python?
    The difference between a function and a method in Python is:
    1. Function: A function is a block of code that is defined independently and can be called directly.
     Ex: len(list)
    2. Method: A method is a function that is associated with an object and is called using the object.
     Ex: list.append()

##2. Explain the concept of function arguments and parameters in Python.
    In Python,parameters and arguments are used to pass data to functions.
    1. Parameters: These are the variables defined in a function's definition that receive values.
    2. Arguments: These are the actual values passed to the function when it is called.

##3. What are the different ways to define and call a function in Python?
    In Python,functions can be defined and called in different ways:
    1. Function with no arguments: Defined without parameters and called directly.
    2. Function with arguments:  Defined with parameters and called by passing values(arguments).
    3. Function with return value: Returns a value using the return statement when called.

##4. What is the purpose of the `return` statement in a Python function?
    The `return` statement in a Python function is used to:
    1. Send a value back to the caller after the function finishes executing.
    2. End the execution of the function immediately when it is encountered.
    3. Allow the result of a function to be reused in expressions,calculations,or stored in variables.

##5. What are iterators in Python and how do they differ from iterables?
    Iterators in Python are objects that are used to traverse through elements one at a time using next() function.
    Difference between Iterable and Iterator:
    1. Iterable: An object that can be looped over (like list,tuple,string).It has an __iter__() method.
    2. Iterator: An object that actually performs the iteration. It has both __iter__() and __next__() methods.
    3. Reusability: An iterable can create multiple iterators,but an iterator can be used only once.

##6. Explain the concept of generators in Python and how they are defined.
     Generators in Python are special functions used to generate values one at
     a time instead of storing all values in memory.
     They are defined using the yield keyword instead of return.
     Key point:
     1. Generators produce values lazily,meaning values are created only when needed.
     2. They help in saving memory,especially for large data sets.
     3. Once a value is yielded,the function's state is saved and resumes from there on the next call.

##7. What are the advantages of using generators over regular functions?
    Advantages of using generators over regular functions:
    1. Memory efficient: generators produce values one at a time instead of storing all results in memory.
    2. Faster for large data sets: values are generated only when required (lazy evaluation).
    3. Simple code for iteration: generators make it easy to write and manage iterative logic.

##8. What is a lambda function in Python and when is it typically used?
    A lambda function in Python is a small anonymous function defined using the lambda keyword.
    Key points:
    1. It can have any number of arguments but only one expression.
    2. It is written in a single line without a function name.
    3. It is typically used for short,simple operations,often with functions like map(),filter(),and sorted().

##9. Explain the purpose and usage of the `map()` function in Python.

    Purpose and Usage:
    1. It applies a function to every element of a list,tuple,etc.
    2. It returns a map object(iterator),not a list directly.
    3. It is commonly used with lambda functions for concise code.

##1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even
```

```python
result = sum_even_numbers([1, 2, 3, 4, 5, 6])
print(result)
```

##4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num <= 1 or num > 200:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
print(is_prime(29))
```

##5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a=0
        self.b=1
        self.count=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value
```

```python
fib = Fibonacci(5)
for num in fib:
    print(num)
```

##6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for power in power_of_two(5):
    print(power)
```

##7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
import os

# Create a dummy data.txt file for demonstration
file_content = """This is line 1.
This is line 2.
And this is the third line.
"""

with open('data.txt', 'w') as f:
    f.write(file_content)

print("Sample 'data.txt' created successfully.")
```

```python
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line

print("Content of data.txt:")
for line in read_file("data.txt"):
    print(line.strip())
```

##8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
data = [(1, 5), (3, 2), (2, 8), (4, 1)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

##9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print(fahrenheit_temperatures)
```

##10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
text = "Hello, World!"
def remove_vowels(text):
    return text.lower() not in "aeiou"

filtered_text = ''.join(filter(remove_vowels, text))
print(filtered_text)
```
